```python
import pandas as pd

df =pd.read_csv('/content/amazon-product-reviews/ratings_Electronics (1).csv', names=['User ID', 'Item ID', 'Rating', 'Timestamp'])
```

## Get only sample of 0.5% only from 7.8M revies

```python
sample = int(0.005 * len(df))
```

```python
df_sample = df.iloc[:sample,:]
```

```python
df_sample.shape
```

## Check missing value usein `isnull` method

```python
df_sample.isnull().sum()
```

### Check mean median mode and 1st qurtile and 3rd qurtile using `describe` method

```python
df_sample.describe()
```

### select only Rating above 3 for build recommendation model

```python
filtered_df = df_sample[df_sample['Rating'] >= 3.0]
```

### Build a Pivot table using `df.pivot_table` and set index as UserID , values as Rating and columns as ItemID

```python
user_ratings = filtered_df.pivot_table(index='User ID', columns='Item ID', values='Rating', fill_value=0)
```

```python
user_ratings
```

### Use `Cosine_similarity` from sklearn and pass through a `Pivot_table`

```python
from sklearn.metrics.pairwise import cosine_similarity

# Similarity Calculation
user_similarity = cosine_similarity(user_ratings)
```

```python
user_ratings.index
```

```python
# Example usage: Print top 5 similar users for user with
user_id = 'A0402564TCEO67AUZFJO'
user_index = user_ratings.index.get_loc(user_id)  # Get the index of the user ID
similar_users = user_similarity[user_index].argsort()[::-1][1:6]  # Exclude the given user itself and get top 5 similar users
similar_users = [user_ratings.index[user] for user in similar_users if user_ratings.index[user] in user_ratings.index]
print("Top 5 similar users for user", user_id, ":")
for user in similar_users:
    print(user)
```
